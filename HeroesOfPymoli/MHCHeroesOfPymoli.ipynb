```python
# Dependencies and Setup

import pandas as pd

#Load File

pymoli_file="../Resources/purchase_data.csv"

pymolidf=pd.read_csv(pymoli_file)

pymolidf.head()
```

```python
#check to see if any rows with missing data
pymolidf.count()
```

```python
#Question 1: Unique Player Count-cf 1.11, rows 3-4

# Calculate the number of unique players (SN) in the dataframe
player_count=len(pymolidf["SN"].unique())

# Place all of the data found into a summary DataFrame
playerdf=pd.DataFrame({"Total Players": [player_count]})

playerdf
```

```python
#Question 2: Purchasing Analysis (Total)

item_count=len(pymolidf["Item ID"].value_counts())
purchase_count=len(pymolidf["Purchase ID"])
revenue_count=(pymolidf["Price"].sum())
average_price=revenue_count/purchase_count

# Place all of the data found into a summary DataFrame

PATdf=pd.DataFrame({"Number of Unique Items": [item_count], "Average Price":[average_price],"Number of Purchases":[purchase_count], "Total Revenue":[revenue_count]})

PATdf


```

```python
#Question 3: Gender Demographics...not total players, but unique players!

# dropping duplicates by SN
pymolidfdropduplicates=pymolidf.drop_duplicates(subset=["SN"])

groupgenderdf=(pymolidfdropduplicates["Gender"].value_counts()) #doesnt necessary need to be groupby

genderplaypct=groupgenderdf/player_count

#on gender: we get 780
#on SN, I think we get 576, but looks weird

groupgenderdf

```

```python
playerdf=pd.DataFrame({"Total Count": groupgenderdf, "Percentage of Players": genderplaypct})
playerdf

```
